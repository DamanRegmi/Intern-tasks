```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data=pd.read_csv('Sample - Superstore.csv',encoding="latin1")
```

```python
customer_columns=['Customer ID','Customer Name','Segment']
product_columns=['Product ID','Category','Sub-Category','Product Name']
order_column=['Order ID','Order Date', 'Ship Date']
shipmode_columns = ['Ship Mode']
sales_columns=['Customer ID', 'Product ID', 'Order ID', 'Ship Mode','Sales', 'Quantity', 'Discount', 'Profit']
```

```python
customer_table = data[customer_columns].drop_duplicates()
product_table=data[product_columns].drop_duplicates()
order_table=data[order_column].drop_duplicates()
shipmode_table=data[shipmode_columns].drop_duplicates()
sale_table=data[sales_columns].drop_duplicates()
```

```python
customer_table.to_csv('cutomer.csv',index=False)
product_table.to_csv('product.csv', index=False)
order_table.to_csv('order.csv', index=False)
shipmode_table.to_csv('shipmode.csv', index=False)
sale_table.to_csv('sale.csv',index=False)
```

```python
from graphviz import Digraph

# Create a Graphviz Digraph for the ERD
erd = Digraph('ERD', format='png')

# Add nodes for each table
erd.node('Customer', 'Customer\n(Customer ID, Customer Name, Segment)')
erd.node('Product', 'Product\n(Product ID, Category, Sub-Category, Product Name)')
erd.node('Order', 'Order\n(Order ID, Order Date, Ship Date)')
erd.node('ShipMode', 'ShipMode\n(Ship Mode)')
erd.node('Sales', 'Sales\n(Customer ID, Product ID, Order ID, Ship Mode, Sale, Quantity, Discount, Profit)')

# Define relationships between tables
erd.edge('Customer', 'Sales', label='Customer_ID')
erd.edge('Product', 'Sales', label='Product_ID')
erd.edge('Order', 'Sales', label='Order_ID')
erd.edge('ShipMode', 'Sales', label='Ship_Mode')

# Render the ERD to a file
erd.render('normalized_erd')
print("ERD saved as normalized_erd.png")
```
