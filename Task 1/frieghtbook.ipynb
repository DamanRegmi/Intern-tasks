```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
baseurl="https://m.freightbook.net"
```

```python
all_link=[]
for x in range(1,249): 
    main=requests.get(f"https://m.freightbook.net/member/results?country={x}")
    s=BeautifulSoup(main.text,"html")
    col=s.find_all("div",class_='link_row text-center')
    
    for column in col:
        link=column.find("a")['href']
        all_link.append(baseurl+link)
print(len(all_link))
```

```python
all_details=[]
for link in all_link:
    r = requests.get(link)
    soup=BeautifulSoup(r.text,"html")
    name=soup.find('h2').text
    try:
        address=soup.find('div',class_='col-xs-7 value street-address').text
    except:
        address="no address"    
    country=soup.find('div',class_='col-xs-7 value country-name').text
    tel=soup.find('div',class_='col-xs-7 value').text
    fax=soup.find('div',class_='col-xs-7 value').text
    details={
        'Name':name,
        'Address':address,
        'Country':country,
        'Tel':tel,
        'Fax':fax
    }
    all_details.append(details)
print(all_details)    
```

```python
df=pd.DataFrame(all_details)
# print(df)
```

```python
df.to_json("frieghtdata.json",orient='records',indent=4)
```
